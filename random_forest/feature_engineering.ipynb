```python
import pandas as pd

def compute_correlation_matrix(df):
    """
    Compute the correlation matrix of the given DataFrame.
    
    Parameters:
    df (pd.DataFrame): Input DataFrame.
    
    Returns:
    pd.DataFrame: Correlation matrix of the DataFrame.
    """
    # Compute the correlation matrix
    correlation_matrix = df.corr()
    return correlation_matrix


import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_matrix(correlation_matrix, figsize=(12, 10)):
    """
    Plot the heatmap of the correlation matrix.
    
    Parameters:
    correlation_matrix (pd.DataFrame): Correlation matrix.
    figsize (tuple): Size of the figure for the plot.
    """
    plt.figure(figsize=figsize)
    sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f', 
                vmin=-1, vmax=1, linewidths=0.5, cbar_kws={'shrink': 0.8})
    plt.title('Correlation Matrix Heatmap')
    plt.savefig('outputs/plots/SQQQ_correlation_matrix.png', bbox_inches='tight')
    plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = compute_correlation_matrix(df)

# Plot the correlation matrix
plot_correlation_matrix(corr_matrix, figsize=(16, 14))
```
